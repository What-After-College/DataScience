# Matplotlib

A Plot is a way of representing data visually. It can describe relationships between variables. Plots can give insight and information about the data that may not have been easily derived by looking at the data as a table.

```python
import pandas as pd
import numpy as np
import pylab as plt
import matplotlib.pyplot as plt
```

### Plotting Linear and Quadratic Equations in Matplotlib

```python
x = np.linspace(0, 2, 10)

plt.plot(x, x, 'o-', label='linear')
plt.plot(x, x ** 2, 'x-', label='quadratic')

plt.legend(loc='best')
plt.title('Linear vs Quadratic progression')
plt.xlabel('Input')
plt.ylabel('Output');
plt.show()
```

```python
import numpy as np
n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Modi ji', 'Kejriwal', 'Rahul', 'Mamta ji'] # list of all the labels
sizes = [38.4, 40.6, 20.7, 10.3] # values of all the labels(vote % of the leaders)
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral'] # list of colors
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90) # we use this fucntion to plot our pie plot by giving all the required parametrs.
plt.title('Vote % of leaders',size=20)
plt.legend(patches, labels, loc="best") # setting up the legend
plt.axis('equal')
plt.tight_layout()
plt.show() # this again shows us the plot
```

```python
# Creatign seperate figures and axis
fig, ax = plt.subplots(figsize = (15, 9))


# Accessign 2nd plot and customizing it.
plot2 = ax.plot(np.arange(1,50,0.1), np.sin(np.arange(1,50,0.1)), c = "Green")
plot2_1 = ax.plot(np.arange(1,50,0.1), np.cos(np.arange(1,50,0.1)), c = "Red")

ax.set(title = "Sin Graph", xlabel = "X Values", ylabel="Sin Values")
ax.legend(("Sin", "Cos"), loc = "upper right")

plt.show()
```

```python
# we will be ploting a bar plot the number of students studying programming languages

fig = plt.figure() 
ax = fig.add_axes([0,0,1,1]) 
langs = ['C', 'C++', 'Java', 'Python', 'PHP'] # these are the x axis values(the languages)
students = [23,17,35,29,12]  # y axis values (no. of students studying each language)
ax.bar(langs,students) # this plots a bar plot for langs and students
plt.title('students vs languages plot',size=20)
plt.xlabel('Languages',size=20)
plt.ylabel('no. of students',size=20)
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv('sales_data.csv')
# plot all numerical columns
df.plot(kind = 'density',subplots = True, layout = (3,3), sharex = False, sharey = False)
# display the plot
plt.show()
```

```python
df.plot(kind = 'box',subplots = True, layout = (3,3), sharex = False, sharey = False)
# display the plot
plt.show()
```

```python
df.plot(kind = 'scatter',x = 'PRICEEACH',y = 'QUANTITYORDERED')
#display plot
plt.show()
```

```python
new_df = df.groupby('PRODUCTLINE').sum()

# Draw a bar plot
new_df.plot(kind = 'bar',y = 'SALES')
plt.show()
```

```python
grouped_df = df.groupby(['YEAR_ID','PRODUCTLINE']).sum()

# plot scatter plot  
grouped_df.unstack().plot(kind = 'bar',y = 'SALES')

#display plot
plt.show()
```
